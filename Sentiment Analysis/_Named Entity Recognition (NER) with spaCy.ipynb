```python
# Install spaCy and its English model
!pip install spacy

# Import spaCy and load the English language model
import spacy
nlp = spacy.load("en_core_web_sm")  # Load the small English model
```

```python
# Define sample text data
text_data = [
    "MetaBrains is expanding its operations to Berlin, Germany, in 2025.",
    "Dr. Sarah Blake, the CTO of MetaBrains, announced the new AI project in New York.",
    "The Great Wall of China is one of the most famous landmarks in the world.",
    "MetaBrains secured a $50 million investment from Quantum Ventures last month.",
]
```

```python
# Process the text data through spaCy
for text in text_data:
    doc = nlp(text)  # Pass the text through the NLP pipeline
    print(f"Text: {text}")
    print("Entities:")
    for ent in doc.ents:
        print(f"  - {ent.text} ({ent.label_})")  # Print the entity and its label
    print("\n")
```

```python
from spacy import displacy

# Visualize entities in a specific sentence
doc = nlp("MetaBrains is expanding its operations to Berlin, Germany, in 2025.")
displacy.render(doc, style="ent", jupyter=True)
```

```python
from spacy.pipeline import EntityRuler

# Add an EntityRuler to the pipeline
ruler = nlp.add_pipe("entity_ruler", before="ner")

# Define custom patterns
patterns = [
    {"label": "PROJECT", "pattern": "MetaBrains AI Project"},
    {"label": "INVESTOR", "pattern": "Quantum Ventures"}
]

# Add the patterns to the EntityRuler
ruler.add_patterns(patterns)

# Test the updated pipeline
doc = nlp("Quantum Ventures invested in the MetaBrains AI Project last month.")
for ent in doc.ents:
    print(f"  - {ent.text} ({ent.label_})")
```

```python
from collections import Counter

# Collect entity types
entity_labels = []
for text in text_data:
    doc = nlp(text)
    entity_labels.extend([ent.label_ for ent in doc.ents])

# Count occurrences of each entity type
entity_counts = Counter(entity_labels)
print("Entity Counts:", entity_counts)
```
