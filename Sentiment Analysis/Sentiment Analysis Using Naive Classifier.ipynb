```python
# Step 1: Import Libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Prepare the Dataset
# Sample product reviews dataset
reviews = [
    "This product is amazing! It exceeded all my expectations.",
    "Absolutely terrible! I will never buy this again.",
    "The quality is outstanding. I highly recommend it.",
    "Worst purchase I've ever made. It broke within a week.",
    "Great value for money. I'm very happy with it.",
    "Completely useless. Don't waste your money.",
    "I'm impressed with how durable this is.",
    "Terrible customer service. Very disappointing experience."
]

# Corresponding labels
sentiments = ['Positive', 'Negative', 'Positive', 'Negative', 'Positive', 'Negative', 'Positive', 'Negative']

# Step 3: Preprocess the Data
# Vectorize the text data
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(reviews)  # Convert text into a numerical representation
```

```python
# Step 4: Split the Dataset
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, sentiments, test_size=0.25, random_state=42)

# Step 5: Train the Model
# Train the model
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# Predict the test data
y_pred = model.predict(X_test)

# Step 7: Evaluate the Model
# Accuracy and classification report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Positive', 'Negative'], yticklabels=['Positive', 'Negative'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Sentiment')
plt.ylabel('Actual Sentiment')
plt.show()
```
