```python
# Step 2: Prepare the Dataset

# Sample dataset of customer reviews
data = {
    "reviews": [
        "The camera quality of this phone is excellent and the battery life is amazing.",
        "The laptop has poor performance but the screen is great for watching movies.",
        "I love the design of the tablet. It is sleek and lightweight.",
        "The delivery was late, and the product packaging was damaged.",
        "The sound quality of these headphones is fantastic!",
        "The software is buggy and crashes frequently.",
        "Amazing customer service and very quick responses to queries.",
        "The build quality is very cheap and feels flimsy."
    ]
}

# Load the data into a pandas DataFrame
df = pd.DataFrame(data)

# Display the dataset
print(df.head())
```

```python
# Step 3: Preprocess the Text Data

# Define stop words
stop_words = stopwords.words('english')

# Preprocessing function
def preprocess(text):
    text = text.lower()                                              # Convert to lowercase
    words = text.split()                                             # Tokenize into words
    words = [word for word in words if word not in stop_words]      # Remove stopwords
    return " ".join(words)

# Apply preprocessing to the reviews column
df['cleaned_reviews'] = df['reviews'].apply(preprocess)

# Display preprocessed data
print(df[['reviews', 'cleaned_reviews']])
```

```python
# Step 4: Convert Text Data into a Document-Term Matrix

# Convert text data into a document-term matrix
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
X = vectorizer.fit_transform(df['cleaned_reviews'])

# Display vocabulary
print("Vocabulary:", vectorizer.get_feature_names_out())

# Display document-term matrix
print("Document-Term Matrix:")
print(X.toarray())
```

```python
# Step 5: Train the LDA Model

# Initialize and train the LDA model
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)

# Display topics
print("Topics discovered:")
for index, topic in enumerate(lda.components_):
    print(f"Topic {index + 1}:", [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-5:]])
```

```python
# Step 6: Visualize Topics

# Install the correct version of pyLDAvis
get_ipython().system('pip install pyLDAvis==3.4.1')

import pyLDAvis
import pyLDAvis.lda_model  # submodule for sklearn models

# Visualize the LDA model
pyLDAvis.enable_notebook()
lda_visualization = pyLDAvis.lda_model.prepare(lda, X, vectorizer)  # Prepare the visualization
pyLDAvis.display(lda_visualization)  # Display the visualization
```
